**12.d. Power3()**

```python
def power3(x, a):
    return pow(x, a)
```

```python
def fx(x):
    return power3(x, 2)
```

```python
df = pd.DataFrame()
for i in range(1,11):
    df = df.append([[i, fx(i)]], ignore_index = True)
```

```python
df.columns = [['x', 'fx']]
```

```python
df
```

```python
plt.xkcd()
plt.figure(figsize = (25, 10))
plt.plot(df['fx'],color='green', linestyle='dashed', marker='o',
         markerfacecolor='yellow', markersize=10)
plt.title('fx vs x')
plt.xlabel('x')
plt.ylabel('fx = x^2')
```

```python
def PlotPower(x1, x2, a):
    df1 = pd.DataFrame()
    for i in range(x1, x2+1):
        df1 = df1.append([[i, power3 (i, a)]], ignore_index = True)
    df1.columns = [['x', 'x^a']]
    plt.xkcd()
    plt.figure(figsize = (25, 10))
    plt.plot(df1[['x^a']])
    plt.xticks(df1.index,df1["x"].values)
    plt.title('x^a vs x')
    plt.xlabel('x')
    plt.ylabel('x^a')
    return df1
```

```python
PlotPower(12, 25, 3)
```
